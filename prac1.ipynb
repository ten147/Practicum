```python
dataframe = pd.read_csv("df1.csv", header=None)
dataframe.dropna(inplace=True)
dataset = dataframe.values
X = dataset[1:,1:16]
Y = dataset[1:,-1]
print(X.shape)
print(Y.shape)
```

```python
Xtrain=X[:-20]
Xtest=X[-20:]
Ytrain=Y[:-20]
Ytest=Y[-20:]
```

```python
#neural network model
def model1(optimizer='adam', init='normal'):
    model = Sequential()
    model.add(Dense(30, input_dim=15, kernel_initializer=init, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2,))
    model.add(Dense(15, kernel_initializer=init, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(1, kernel_initializer=init, activation='relu'))
    model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])
    return model
```

```python
model=model1()
```

```python
model.fit(Xtrain,Ytrain,epochs=200, batch_size=10, verbose=2)
```

```python
ppred=model.predict(Xtest)
```

```python
print(Ytest)
print(ppred)
```

```python
from sklearn.metrics import r2_score
r2_score(Ytest, ppred)
```

```python
model.evaluate(Xtest,Ytest)
```
